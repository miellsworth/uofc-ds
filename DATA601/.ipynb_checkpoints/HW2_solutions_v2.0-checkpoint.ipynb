## Question 1

### Text frequency analysis

This question asks you to perform a simple frequency analysis on the novel "Moby Dick" by _Herman Melville_. To simplify things, the code cell below reads the novel from a text file, splits it into words removing any punctuation and empty strings. The resulting words are stored in a list (called `book`). It also reads in another text file into a list (called `top`) consisting of the 100 most frequently occuring words in the English language. 

Please execute the code cell and then perform the following tasks. _You may use any built in data structures or functions_.

- For each of the words in `top`, determine how many times the word appears in `book`. Keep a record of this tally in a suitable data structure.
- Rank your tallied results; the most frequent word gets rank 1, the second most frequent gets rank 2, and so on.
- Plot a rank-frequency distribution, with rank on the horizontal axis and frequency on the vertical axis.
- What kind of a trend do you observe and how does it compare to [Zipf's law](https://simple.wikipedia.org/wiki/Zipf%27s_law)?


```python
import re

def fileToList( fname, regexp=r'\W+' ):
    file = open(fname, 'rt')
    text = file.read()
    file.close()
    # split based on provided regular expression and remove empty strings
    # By default, matches words.
    return [x for x in re.split(regexp, text) if x]
    
book = fileToList( 'moby_dick.txt' ) # list of words in the novel
top = fileToList( 'top100.txt' ) # Top 100 words
    
print(book[10000:10100])
print("\n")
print(top)
print("\n")
print(len(top))
```

## Question 1 *Solution*

```python
# normalize all words to be uppercase
book = [word.upper() for word in book]

# count occurences of top words in book
freqs = [(word,book.count(word.upper())) for word in top]
print(freqs)
print('\n')

# rank tallied results
freqs.sort(key = lambda x: x[1], reverse = True)
ranked = [(i,word[0]) for i,word in zip(range(1,len(freqs)+1),freqs)]
print(ranked)

# calculated predicted frequency based on Zipf's law
zipf_predicted = [freqs[0][1]*(1/N) for N in range(1,len(freqs)+1)]

# plot data
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.plot( range(1,101), [x[1] for x in freqs], 'ko')
plt.plot( range(1,101), zipf_predicted, 'b-')
# plt.yscale('log')
fig.set_size_inches(15,6)
plt.legend(('actual data', 'zipf prediction'))

plt.xlabel('rank')
plt.ylabel('frequency (count)')
plt.suptitle('Rank-frequency Distribution')
plt.show()
```

The data fits the Zipf law model approximately. See above graph.

## Question 2:

### Visualizing trends in an index

This question asks you to plot the Bitcoin price index (BPI) along with 5-day and 10-day averages. Please execute the code cell below; it will read in the daily closing price from Sep. 1, 2017 to Aug. 30, 2018 (data obtained from https://www.coindesk.com/price/). Perform the following tasks.

- Observe that the closing prices are at a daily interval. We therefore do not need the date information. Clean up the data and only retain the price information. Store the result in a list or a numpy array in floating point format.
- Recall that a [simple moving average](https://en.wikipedia.org/wiki/Moving_average#Simple_moving_average) is defined as the (unweighted) mean over the previous $N$ days.<br> 
Define a function or a generator to perform a simple moving average of the price index. The number of days $N$ to average over should be an input parameter to the function/generator.  
- Plot the raw price index data along with 5-day and 10-day simple moving average. Plot on the same figure in order to help you visually ascertain the effect of the filter.
- What is the effect of the moving average filter? In what circumstances would you _not_ want to use a moving average?

```python
bfile = "coindesk-bpi-USD-close_data.csv"
bpi = fileToList( bfile, regexp=r'[,\r\n]+' )
print(bpi[:10:1])
print(bpi[-10::1])
```

## Question 2 *Solution*

```python
import numpy as np

# filter out timestamps
bpindex = np.array([float(x) for x in bpi[1::2]])

# define a generator that yeilds, consecutive, N day slices of the data array
def winitr(arr, N):
    s = 0
    while (s+N) <= arr.size:
        yield arr[s:s+N:1]
        s += 1
        
# build simple moving averages
five_day = np.array([win.mean() for win in winitr(bpindex, 5)])
ten_day = np.array([win.mean() for win in winitr(bpindex, 10)])

# append NaNs to front of lists to indicate omitted data where there isn't enough prior data 
five_day = np.concatenate( (np.array([np.nan]*4), five_day) )
ten_day = np.concatenate( (np.array([np.nan]*9), ten_day) )
            
        
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.plot(bpindex, 'k-', alpha=0.25)
plt.plot(five_day, 'g-')
plt.plot(ten_day, 'b-')
plt.ylabel('BTC price')
plt.xlabel('day')
plt.suptitle('Bitcoin Prices')
plt.legend(('raw data', '5-day SMA', '10-day SMA'))
fig.set_size_inches(15,6)
```

The moving average filter has a smoothing effect. A smoothing effect is not 
useful when high frequency variations constitute important information,
e.g., high frequency financial trading.

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import networkx as nx
import seaborn as sns
```

We'll be playing with a dataset of Marvel superheros, connected by their co-occurrence in the comics.
Originally derived from a 2011 InfoChimps data dump. Since published elsewhere. http://syntagmatic.github.io/exposedata/marvel/

```python
# Load the dataset and save it as a graph
marvel_all_pairs = pd.read_csv(r'Marvel-AllPairs.tsv', sep="\t") 
# Load the dataset and save it as a graph
marvel_50plus_pairs = pd.read_csv(r'Marvel-50plus.tsv', sep="\t") 
# Let's start with only the characters who have appeared together 250+ times
marvel_250plus_pairs = marvel_50plus_pairs[marvel_50plus_pairs["Co-Featured"] > 250] 

def marvel_pairs_to_graph(input_nodes):
    new_graph = nx.Graph()
    for row in input_nodes.iterrows(): 
        new_graph.add_node(row[1][0]) # Add the first character
        new_graph.add_node(row[1][1]) # Add the second character
        new_graph.add_edge(row[1][0],row[1][1],cofeatured=row[1][2])
    return new_graph
    
marvel_250plus_graph = marvel_pairs_to_graph(marvel_250plus_pairs)
marvel_50plus_graph = marvel_pairs_to_graph(marvel_50plus_pairs)
marvel_all_graph = marvel_pairs_to_graph(marvel_all_pairs)
```

```python
nx.draw(marvel_250plus_graph, with_labels=True)
```

```python
# Calculating measures of centrality (Remember that you can use Pandas and NetworkX together).
degree_centrality_250plus = pd.DataFrame(list(nx.degree_centrality(marvel_250plus_graph).items()))
degree_centrality_250plus.sort_values(1,ascending=False).head(10)
```

```python
# NetworkX also lets you do cool things like find shortest paths in the graph.
# (hint - it also has methods for slicing graphs in lots of other useful ways)
nx.shortest_path(marvel_250plus_graph, 'SHADOWCAT','MARVEL GIRL')
```

**Question 2.** 

**(A)** What is the largest connected component in the Marvel_250plus graph? What fraction of the characters in the dataset does it contain?

**(B)** What about the largest connected component in the Marvel_50plus graph?

**(C)** What about the largest connected component in the Marvel_AllPairs graph? *(This may take a while!)*

**(D)** Can you explain the difference?

```python
# 2
# (A)
print([len(c) for c in sorted(nx.connected_components(marvel_250plus_graph), key=len, reverse=True)])
largest_cc_250 = max(nx.connected_components(marvel_250plus_graph), key=len)
largest_cc_250_size = len(largest_cc_250)
count_marvel_characters_250 = len(marvel_250plus_pairs.Source.append(marvel_250plus_pairs.Target).unique())
print(largest_cc_250_size)
print(count_marvel_characters_250)
print(largest_cc_250_size/count_marvel_characters_250)

# (B)
print([len(c) for c in sorted(nx.connected_components(marvel_50plus_graph), key=len, reverse=True)])
largest_cc_50 = max(nx.connected_components(marvel_50plus_graph), key=len)
largest_cc_50_size = len(largest_cc_50)
count_marvel_characters_50 = len(marvel_50plus_pairs.Source.append(marvel_50plus_pairs.Target).unique())
print(largest_cc_50_size)
print(count_marvel_characters_50)
print(largest_cc_50_size/count_marvel_characters_50)

# (C)
print([len(c) for c in sorted(nx.connected_components(marvel_all_graph), key=len, reverse=True)])
largest_cc_all = max(nx.connected_components(marvel_all_graph), key=len)
largest_cc_all_size = len(largest_cc_all)
count_marvel_characters_all = len(marvel_all_pairs.Source.append(marvel_all_pairs.Target).unique())
print(largest_cc_all_size)
print(count_marvel_characters_all)
print(largest_cc_all_size/count_marvel_characters_all)
```

**Question 3.** Take the *largest connected component of Marvel-50plus graph* (which you should have found in Question 2) and divide into communities using at least two community detection algorithms and plot them. Which community-detection algorithm and parameters gives you the most satisfying split?

```python
# Get the set of communities after n splits
marvel_communities = nx.community.girvan_newman(Y) 
n = 7
for i in range(1,n):
    n_communities = next(marvel_communities)
n_communities

# Visualize the community detection result using force-directed layout

# Create spring layout, and save it as "pos". Use nx.spring_layout(g).
pos = nx.kamada_kawai_layout(Y)


# Plot the nodes of each community using different colors.
palette = sns.color_palette('hls', len(n_communities))
fig, ax = plt.subplots(figsize=(18,20))
for i, ith_community in enumerate(n_communities):
    nx.draw_networkx_nodes(Y, pos, ith_community, 
                           node_size = 200,
                           font_size = 10,
                           node_color = palette[i], 
                           ax=ax)

'''Lesson: To draw edges and labels, use 
           nx.draw_networkx_edges(g) and
           nx.draw_networkx_labels(g).'''

# Draw network edges and labels
nx.draw_networkx_labels(Y, pos)
nx.draw_networkx_edges(Y, pos)

# Add a title "Community Detection and Force-directed Layout" using ax.set_title()
ax.set_title("Community Detection with Girvan Newman and Kamada Kawai Layout", fontdict={'fontsize': 20, 'fontweight': 'medium'})
```

```python
# Get the set of communities after n splits
marvel_communities = list(nx.community.greedy_modularity_communities(Y))

# Visualize the community detection result using kamada kawai layout

# Create kamada kawai layout, and save it as "pos". Use nx.kamada_kawai_layout(g).
pos = nx.kamada_kawai_layout(Y)


# Plot the nodes of each community using different colors.
palette = sns.color_palette('hls', len(marvel_communities))
fig, ax = plt.subplots(figsize=(18,20))
for i, ith_community in enumerate(marvel_communities):
    nx.draw_networkx_nodes(Y, pos, ith_community, 
                           node_size = 200,
                           font_size = 10,
                           node_color = palette[i], 
                           ax=ax)

'''Lesson: To draw edges and labels, use 
           nx.draw_networkx_edges(g) and
           nx.draw_networkx_labels(g).'''

# Draw network edges and labels
nx.draw_networkx_labels(Y, pos)
nx.draw_networkx_edges(Y, pos)

# Add a title "Community Detection and Force-directed Layout" using ax.set_title()
ax.set_title("Community Detection with Greedy Modularity and Kamada Kawai Layout", fontdict={'fontsize': 20, 'fontweight': 'medium'})
```

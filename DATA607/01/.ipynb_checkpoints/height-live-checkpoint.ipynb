```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load data.
df = pd.read_csv("./height.csv")
df
```

```python
# Parametric estimate of densities
# class-conditional heights are normally distributed

MALE = 0
FEMALE = 1

X_m = df[df.Gender == "Male"].Height.values
mean_m = X_m.mean()
std_m = X_m.std()

X_f = df[df.Gender == "Female"].Height.values
mean_f = X_f.mean()
std_f = X_f.std()

X = df.Height.values
```

$$
g(x) = \frac{1}{\sqrt{2\pi}\sigma}\exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)
$$

```python
def g(x, mean, std):
    return np.exp(-(x - mean)**2/(2*std**2))/(np.sqrt(2*np.pi)*std)
```

```python
np.min(X), np.max(X)
```

```python
x = np.linspace(50, 80, 100);
y_m = g(x, mean_m, std_m)
y_f = g(x, mean_f, std_f)
w_m = len(X_m)/(len(X_m) + len(X_f)) # mixture proportions
w_f = len(X_f)/(len(X_m) + len(X_f))
y = w_m*y_m + w_f*y_f
plt.plot(x, y_m)
plt.plot(x, y_f)
plt.plot(x, y, lw=3)
```

```python
## Histogram estimators of heights

h, bins = np.histogram(X, bins=20, density=True)

plt.hist(X_m, bins=bins, density=True, ec='k');
plt.hist(X_f, bins=bins, density=True, ec='k');
plt.hist(X, bins=bins, density=True, ec='k');
plt.plot(x, y_m, lw=3)
plt.plot(x, y_f, lw=3)
plt.plot(x, y, lw=3)
```

## Unsupervised Learning

### Mixture of Gaussians

```python
np.random.shuffle(X)
```

```python
from sklearn.mixture import GaussianMixture

M = GaussianMixture(n_components=2)
M.fit(X.reshape(-1, 1))
```

```python
mean_0, mean_1 = M.means_
std_0, std_1 = np.sqrt(M.covariances_).squeeze()
```

```python
plt.plot(x, y_m, lw=3)
plt.plot(x, y_f, lw=3)
plt.plot(x, g(x, mean_0, std_0))
plt.plot(x, g(x, mean_1, std_1))
```

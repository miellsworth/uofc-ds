## Simple Linear Regression

```python
import numpy as np
from matplotlib import pyplot as plt
```

### Model
$$y=\beta_0 + \beta_1x + \epsilon,\qquad \epsilon\sim N(0, \sigma^2)$$

```python
# Seed the pseudorandom number generator, for reproducibility.
np.random.seed(666)

# intercept and slope
intercept = np.random.random()
slope = np.random.random() - 1
print(f"intercept: {intercept:.2}, slope: {slope:.2}")

# standard deviation of error distribution
sigma = 0.025
```

### Sample

```python
# sample size

n = 20
x = np.random.uniform(size=n)
e = np.random.normal(0, sigma, size=n)
y = b0 + b1*x + e

plt.plot(x, y, 'o', mec='k', label="sample")
plt.plot([0, 1], [intercept, intercept + slope*1], '-', lw=3, label="ground truth")
plt.legend();
```

### Fit the model to the data
$$
\widehat\beta = (X^TX)^{-1}X^Ty
$$

```python
X = np.ones((n, 2))
X[:,1] = x

# method 1
intercept_, slope_ = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
print(f"1. fit intercept: {intercept_:.2}, fit slope: {slope_:.2}")

# method 2
intercept_, slope_ = np.linalg.inv(X.T @ X) @ X.T @ y
print(f"2. fit intercept: {intercept_:.2}, fit slope: {slope_:.2}")

# method 3
(intercept_, slope_), *_ = np.linalg.lstsq(X, y, rcond=-1)
print(f"3. fit intercept: {intercept_:.2}, fit slope: {slope_:.2}")

# method 4
from sklearn.linear_model import LinearRegression
M = LinearRegression()
M.fit(x.reshape(-1, 1), y)
print(f"3. fit intercept: {M.intercept_:.2}, fit slope: {M.coef_[0]:.2}")
```

```python
plt.plot(x, y, 'o', mec='k', label="sample")
plt.plot([0, 1], [intercept, intercept + slope*1], '-', lw=3, label="ground truth")
plt.plot([0, 1], [intercept_, intercept_ + slope_*1], '-', lw=3, label="least squares line")
plt.legend();
```

### A custom regressor

```python
from sklearn.base import BaseEstimator, RegressorMixin
from sklearn.utils import check_X_y, check_array

class SLR(RegressorMixin):
    """our custom regressor"""
    
    def fit(self, X, y):
        X, y = check_X_y(X, y)
        XX = np.insert(X, 0, np.ones(len(X)), axis=1)
        XX[:,1] = X[:,0]
        
        intercept_, slope_ = np.linalg.inv(XX.T @ XX) @ XX.T @ y

        setattr(self, "intercept_", intercept_)
        setattr(self, "slope_", slope_)
        
        return self
        
    def predict(self, X):
        X = check_array(X)
        return self.intercept_ + self.slope_*X
```

```python
M = SLR()
M.fit(x.reshape(-1, 1), y)
print(f"fit intercept: {M.intercept_:.2}, fit slope: {M.slope_:.2}")

# inherited from RegressorMixin
print(f"R^2 score: {M.score(x.reshape(-1, 1), y):.2}")
```

### Simple linear regression as a linear smoother
$$
\hat r(x) = \mathbf{w}(x)\cdot \mathbf{y}
$$

```python
u = np.random.uniform(size=10)
U = np.ones((10, 2))
U[:,1] = u
w = U @ np.linalg.inv(X.T @ X) @ X.T
v = w.dot(y)

plt.plot(x, y, 'o', mec='k', label="sample")
plt.plot([0, 1], [intercept, intercept + slope*1], '-', lw=3, label="ground truth")
plt.plot([0, 1], [intercept_, intercept_ + slope_*1], '-', lw=3, label="least squares line")
plt.plot(u, v, 'o', mec='k', ms=10, label="linear smoother")
plt.legend();
```

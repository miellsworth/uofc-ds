```python
# Import packages

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Dense, Dropout, Flatten, Input, MaxPooling2D
from keras.callbacks import ModelCheckpoint
from keras.models import load_model

from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import confusion_matrix

import random
import itertools
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

In class on Wednesday, March 26 we trained two neural networks models for classifying fashion MNIST.
We'll call them **Model A** and **Model B**.

```python
# Model A
modelA = Sequential([Input((28, 28)),
                    Flatten(),
                    Dense(256, activation="relu"),
                    Dropout(0.2),
                    Dense(128, activation="relu"),
                    Dense(64, activation="relu"),
                    Dropout(0.2),
                    Dense(10, activation="softmax")], name="Model_A")
modelA.summary()
modelA.compile(loss="categorical_crossentropy", optimizer="sgd", metrics=["accuracy"])
```

```python
# Model B
modelB = Sequential(name="Model_B")
modelB.add(Conv2D(64, (7, 7), activation="relu", padding="same", input_shape=(28, 28, 1))) # zero-padding
modelB.add(MaxPooling2D(2))
modelB.add(Conv2D(128, (3, 3), activation="relu", padding="same"))
modelB.add(Conv2D(128, (3, 3), activation="relu", padding="same"))
modelB.add(MaxPooling2D(2))
modelB.add(Conv2D(256, (3, 3), activation="relu", padding="same"))
modelB.add(Conv2D(256, (3, 3), activation="relu", padding="same"))
modelB.add(MaxPooling2D(2))
modelB.add(Flatten())
modelB.add(Dense(128, activation="relu"))
modelB.add(Dropout(0.5))
modelB.add(Dense(64, activation="relu"))
modelB.add(Dropout(0.5))
modelB.add(Dense(10, activation="softmax"))
modelB.summary()
modelB.compile(loss="categorical_crossentropy", optimizer="sgd", metrics=["accuracy"])
```

### 1.

**Train** each model for 20 epochs, just like we did in class and then **save** the models. (See [the docs](https://keras.io/getting-started/faq/#how-can-i-save-a-keras-model) or page 314 of G&eacute;ron.)

```python
# Load the MNIST fashion dataset
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# Rescale the image values to a range of 0 to 1
train_images = np.expand_dims((train_images / 255.0).astype(np.float32), axis = 3)
test_images = np.expand_dims((test_images / 255.0).astype(np.float32), axis = 3)

# One Hote Encode
train_labels_OHE = np.zeros((60000, 10)).astype(np.float32)
train_labels_OHE[np.arange(60000), train_labels] = 1

test_labels_OHE = np.zeros((10000, 10)).astype(np.float32)
test_labels_OHE[np.arange(10000), test_labels] = 1

# Shape
print(train_images.shape, test_images.shape, train_labels_OHE.shape, test_labels_OHE.shape)
```

```python
fashion_mnist = keras.datasets.fashion_mnist

(x_tr, y_tr), (x_te, y_te) = fashion_mnist.load_data()

X_tr = (x_tr/255.).astype(np.float32)
Y_tr = np.zeros((60000, 10)).astype(np.float32)
Y_tr[np.arange(60000), y_tr] = 1

X_te = (x_te/255.).astype(np.float32)
Y_te = np.zeros((10000, 10)).astype(np.float32)
Y_te[np.arange(10000), y_te] = 1

CLASSES = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
           "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]

X_tr = np.expand_dims(X_tr, axis=3)
X_te = np.expand_dims(X_te, axis=3)
print(X_tr.shape, X_te.shape, Y_tr.shape, Y_te.shape)
```

```python
# Train and save Model A
modelA.fit(train_images, train_labels_OHE, epochs = 20, batch_size = 32, validation_split = 0.2)
modelA.save('modelA.h5')

# Train and save Model B
modelB.fit(train_images, train_labels_OHE, epochs = 20, batch_size = 32, validation_split = 0.2)
modelB.save('modelB.h5')
```

### 2.

Do this exercise using **either Model A or Model B**.

Construct the confusion matrix $C$ of the test set, i.e., the $10\times 10$ matrix whose $(i,j)$-entry is the number of images of class $i$ mistakenly classified as class $j$.
Produce a nice visualization of the confusion matrix in the spirit of [this example](https://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html) from the scikit-learn docs.

The **symmetric confusion score** for an unordered pair $\{i,j\}$ of indices is the sum $C_{ij} + C_{ji}$. It's the number of times images of classes $i$ and $j$ were mistaken for one another. Which pairs of features have the three highest symmetrized confusion scores?

```python
# Create an array of predicted classes based on the max probability, to be used in the confusion matrix and Question 3
predictions = np.argmax(modelA.predict(test_images), axis = 1)

# Create an array of max proabilities to be used in Question 3
probabilities = np.max(modelA.predict(test_images), axis = 1)

# create an array of probabilities to be used in Question 3
probabilities_all = modelA.predict(test_images)
```

```python
# Confusion matrix plot code adapted from here: https://towardsdatascience.com/a-simple-cnn-multi-image-classifier-31c463324fa

# Create a confusion matrix
cm = confusion_matrix(test_labels, predictions)

# Define a function to plot the confusion matrix, taking the confusion matrix and classes as inputs
def plot_cm(cm, classes, normalize = False, cmap = plt.cm.Blues):
   
    # Option to normalize data
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        norm_flag = ' Normalized'
    else:
        norm_flag = ''

    plt.figure(figsize=(12, 12))

    # Plots the colormap of the confusion matrix
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title('Confusion Matrix' + norm_flag)

    # Adds a legend to describe the color gradient
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt), horizontalalignment='center', color='white' if cm[i, j] > thresh else 'black')

    plt.tight_layout()
    plt.ylabel('Test Labels')
    plt.xlabel('Prediction Labels')

# Use the defined function to plot the confusion matrix
plot_cm(cm, class_names, normalize = False)
```

```python
# Calculate symmetric confusion score in an array
C = np.tril(cm, k = -1).T + np.triu(cm, k = 1)

# Take the top 3 indices from the array
C_top3ind = np.unravel_index(np.argsort(C.ravel())[-3:], C.shape)

# Reorder the top 3 indices
C_top3ind = np.flip(C_top3ind, 1)

# Create list image pairs that score highest in the confusion score array
top_3 = []
for i in range(len(C_top3ind[0])):
    top_3.append([class_names[C_top3ind[0][i]], class_names[C_top3ind[1][i]]])
print(top_3)
```

As shown above, the classes "T-shirt / top"  and "Shirt" are most often mistaken, followed by the class pairs "Pullover" and "Coat", and "Pullover" and "Shirt.

For each ordered pair of indices $(i,j)$, find the image of class $i$ mistakenly classified as class $j$ with the lowest/highest
- categorical crossentropy loss is the lowest/highest. Loss functions are located in the `tensorflow.keras.losses` module.

- predicted probability of belonging to class $j$. These predicted probabilities are the output of `model.predict`.

Plot these extremal predictions in grids; four grids in total corresponding to the four possible choices of lowest/highest and loss/probability. Assuming you're using `matplotlib`, I suggest either [`subplots`](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.subplot.html) or [`ImageGrid`](https://matplotlib.org/3.1.1/api/_as_gen/mpl_toolkits.axes_grid1.axes_grid.ImageGrid.html#mpl_toolkits.axes_grid1.axes_grid.ImageGrid).

```python
# Create an array for the indices of the image set where the predictions are incorrect (not equal to test_labels)
incorrect_ind = np.argwhere(test_labels != predictions).flatten()

# Create an array of classes that were predicted incorrectly
incorrect_preds = predictions[incorrect_ind].flatten()

# Create an array of the classes that should have been predicted
actual = test_labels[incorrect_ind].flatten()

# Create an array of images that were predicted incorrectly
incorrect_imgs = X_tr.reshape((60000, 28, 28))[incorrect_ind]

# Calculate the categorical_crossentropy loss from the test labels and probabilities
loss = np.array(keras.losses.categorical_crossentropy(test_labels_OHE, probabilities_all))[incorrect_ind].flatten()

# Create an array of the probabilities of the incorrect predictions
probs = probabilities[incorrect_ind].flatten()

# Combine the arrays into a DataFrame for pulling the maximum and minimum values
df = pd.DataFrame({'prediction':incorrect_preds, 'actual':actual,'loss':loss, 'probabilities':probs, 'index':incorrect_ind})

# Group by each prediction / actual class pair (i, j) and determine the max and minimum values for the loss and probability
max_loss = pd.DataFrame(df.groupby(by=['prediction', 'actual'])['loss'].max().reset_index())
max_prob = pd.DataFrame(df.groupby(by=['prediction', 'actual'])['probabilities'].max().reset_index())
min_loss = pd.DataFrame(df.groupby(by=['prediction', 'actual'])['loss'].min().reset_index())
min_prob = pd.DataFrame(df.groupby(by=['prediction', 'actual'])['probabilities'].min().reset_index())

# Group by each prediction / actual class pair (i, j) and determine the indices of the max and minimum values for the loss and probability
max_loss_ind_df = pd.DataFrame(df.groupby(by=['prediction', 'actual'])['loss'].idxmax().reset_index())
max_prob_ind_df = pd.DataFrame(df.groupby(by=['prediction', 'actual'])['probabilities'].idxmax().reset_index())
min_loss_ind_df = pd.DataFrame(df.groupby(by=['prediction', 'actual'])['loss'].idxmin().reset_index())
min_prob_ind_df = pd.DataFrame(df.groupby(by=['prediction', 'actual'])['probabilities'].idxmin().reset_index())

# Define a function that will use the DataFrames with the indices of the max and minimum values of the loss and probability
# and return an array of indices that can be used for plotting
def pair_ind(ind_df, type):
    result = [[] for i in range(len(CLASSES))]
    for i in range(len(CLASSES)):
        for j in range(len(CLASSES)):
            if j in ind_df[ind_df.prediction == i].actual.values:
            result[i].append(df.iloc[ind_df[(ind_df.prediction == i) & (ind_df.actual == j)][type].values.item(0)]['index'])
        else:
            result[i].append(0)
    return np.array(result)

# Use the previously defined function to create four arrays of indices for plotting in the next code chunk
max_loss_ind = np.array(pair_ind(max_loss_ind_df, 'loss'))
max_prob_ind = np.array(pair_ind(max_prob_ind_df, 'probabilities'))
min_loss_ind = np.array(pair_ind(min_loss_ind_df, 'loss'))
min_prob_ind = np.array(pair_ind(min_prob_ind_df, 'probabilities'))
```

```python
# Define a function that will plot the incorrectly predicted images with the maximum and minimum loss and probaility values
def plot_mistakes(array_ind):
    n_rows = 10
    n_cols = 10
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 12))
    for index, ax in enumerate(axes.flatten()):
        ind = int(array_ind.flatten()[index])
        if ind == 0:
            img = np.zeros((28,28))
        else:
            img = x_te[ind]
            ax.imshow(img, cmap="binary", interpolation="nearest")
        if index < 10:
            ax.set_title(CLASSES[index])
        if index == 0 or index % 10 == 0:
            ax.set_ylabel(CLASSES[index // 10], rotation = 0)
            ax.get_xaxis().set_ticks([])
            ax.get_yaxis().set_ticks([])
        for loc in ['top', 'bottom', 'right', 'left']:
            ax.spines[loc].set_visible(False)
        else:
            ax.axis('off')
    plt.tight_layout()
    plt.show()

# Plot the matrices
plot_mistakes(max_loss_ind)
plot_mistakes(max_prob_ind)
plot_mistakes(min_loss_ind)
plot_mistakes(min_prob_ind)
```

### 4.

We use the architecture and weights of a neural network trained on a large data set as initialization values for retraining the same network on a much smaller data set of a similar nature. This approach is known as **transfer learning**.

Load the **MNIST handwritten digit data** (`keras.datasets.mnist.load_data`), both training and testing sets.
For $n=100, 200, ...$, retrain your models A and B on a $n$ randomly selected training images and record the accuracy on the full test set.
How large does $n$ need to be to achieve 95% accuracy on the test set? Make sure you reload your weights between training runs!

For $n=100, 200, ...$, train models A and B from scratch (i.e., with random initializations) on a $n$ randomly selected training images and record the accuracy on the full test set.
How large does $n$ need to be to achieve 95% accuracy on the test set? Make sure you reinitialize your weights between training runs!

Comment on the effectiveness of transfer learning in the context of your computations.



```python
# Load the MNIST handwritten digit dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Rescale the image values to a range of 0 to 1
x_train = np.expand_dims((x_train / 255.0).astype(np.float32), axis = 3)
x_test = np.expand_dims((x_test / 255.0).astype(np.float32), axis = 3)

# One Hote Encode
y_train_OHE = np.zeros((60000, 10)).astype(np.float32)
y_train_OHE[np.arange(60000), y_train] = 1

y_test_OHE = np.zeros((10000, 10)).astype(np.float32)
y_test_OHE[np.arange(10000), y_test] = 1

# Shape
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
# Define a function that will reinitialize the weights between training runs
def reset_weights(model):
    for layer in model.layers: 
        if isinstance(layer, tf.keras.Model):
            reset_weights(layer)
            continue
    for k, initializer in layer.__dict__.items():
        if "initializer" not in k:
            continue
# find the corresponding variable
    var = getattr(layer, k.replace("_initializer", ""))
    var.assign(initializer(var.shape, var.dtype))
```

```python
# Loop through the pre-trained Model A and stop when accuracy exceeds 0.95
for n in range(100, 60001, 200):
    random_indA_trained = random.sample(range(0, 60000), n)
    print(f'N = {n}')
    modelA_trained = keras.models.load_model('modelA.h5')
    modelA_trained.fit(x_train[random_indA_trained], y_train_OHE[random_indA_trained], epochs = 20, batch_size = 32)
    resultsA_trained = modelA_trained.evaluate(x_test, y_test_OHE)
    print(resultsA_trained)
    if resultsA_trained[1] >= 0.95:
        break
```

```python
# Loop through the pre-trained Model B and stop when accuracy exceeds 0.95
for n in range(100, 60001, 200):
    random_indB_trained = random.sample(range(0, 60000), n)
    print(f'N = {n}')
    modelB_trained = keras.models.load_model('modelB.h5')
    modelB_trained.fit(x_train[random_indB_trained], y_train_OHE[random_indB_trained], epochs = 20, batch_size = 32)
    resultsB_trained = modelB_trained.evaluate(x_test, y_test_OHE)
    print(resultsB_trained)
    if resultsB_trained[1] >= 0.95:
        break
```

```python
# Loop through the un-trained Model A and stop when accuracy exceeds 0.95
for n in range(100, 60001, 200):
    random_indA = random.sample(range(0, 60000), n)
    print(f'N = {n}')
    reset_weights(modelA)
    modelA.fit(x_train[random_indA], y_train_OHE[random_indA], epochs = 20, batch_size = 32)
    resultsA = modelA.evaluate(x_test, y_test_OHE)
    print(resultsA)
    if resultsA[1] >= 0.95:
        break
```

```python
# Loop through the un-trained Model B and stop when accuracy exceeds 0.95
for n in range(100, 60001, 200):
    random_indB = random.sample(range(0, 60000), n)
    print(f'N = {n}')
    reset_weights(modelB)
    modelB.fit(x_train[random_indB], y_train_OHE[random_indB], epochs = 20, batch_size = 32)
    resultsB = modelB.evaluate(x_test, y_test_OHE)
    print(resultsB)
    if resultsB[1] >= 0.95:
        break
```

This is significant enough to show the value of transfer learning for this particular dataset.
